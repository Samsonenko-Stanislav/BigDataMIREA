```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
=data = pd.read_csv('./train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
# Числовые признаки
num_cols = [
    'ClientPeriod',
    'MonthlySpending',
    'TotalSpent'
]

# Категориальные признаки
cat_cols = [
    'Sex',
    'IsSeniorCitizen',
    'HasPartner',
    'HasChild',
    'HasPhoneService',
    'HasMultiplePhoneNumbers',
    'HasInternetService',
    'HasOnlineSecurityService',
    'HasOnlineBackup',
    'HasDeviceProtection',
    'HasTechSupportAccess',
    'HasOnlineTV',
    'HasMovieSubscription',
    'HasContractPhone',
    'IsBillingPaperless',
    'PaymentMethod'
]

feature_cols = num_cols + cat_cols
target_col = 'Churn'
```

```python
y = data['Churn']
data = data.drop(columns=['Churn'])
```

```python
data.isna().sum()
```

```python
data['IsSeniorCitizen'] = data['IsSeniorCitizen'].replace(0, 'No')
data['IsSeniorCitizen'] = data['IsSeniorCitizen'].replace(1, 'Yes')
```

```python
data.replace(" ", np.nan, inplace=True)
data['TotalSpent'] = data['TotalSpent'].fillna(0).astype(float)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
data.columns = [str(t) for t in data.columns]
X_train, X_valid, y_train, y_valid = train_test_split(data, y, train_size=0.8, random_state=42)
```

```python
from sklearn.compose import make_column_transformer, ColumnTransformer
# отнормируем числовые признаки и закодируем категориальные
preprocessor = make_column_transformer(
    (make_pipeline(StandardScaler(), RobustScaler()), num_cols),
    (OneHotEncoder(), cat_cols)
)
```

```python
# Реализовать баггинг.
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
```

```python

random_forest = RandomForestClassifier()
params_grid = {
    "max_depth": [9, 15, 18, 24],
    "min_samples_leaf": [1, 2, 3],
    "min_samples_split": [8, 10, 12]
}

grid_search_random_forest = GridSearchCV(
    estimator=random_forest,
    param_grid=params_grid,
    scoring="f1_macro",
)
```

```python
model = make_pipeline(
    preprocessor,

  grid_search_random_forest,
)
```

```python
import time
```

```python
start = time.time()
model.fit(X_train, y_train)
end = time.time()
print("Время обучения:", end - start)
```

```python
model
```

```python
best_forest = grid_search_random_forest.best_params_

best_forest
```

```python
best_forest_pred_train =  model.predict(X_train.copy())
best_forest_f1_score_train = f1_score(best_forest_pred_train, y_train.copy(), average='macro')
print(f'F1-score train: {best_forest_f1_score_train}')

best_forest_pred_test = model.predict(X_valid.copy())
best_forest_f1_score_test = f1_score(best_forest_pred_test, y_valid.copy(), average='macro')
print(f'F1-score valid: {best_forest_f1_score_test}')
```

```python
import catboost as cb
```

```python
model_catboost_clf = cb.CatBoostClassifier(
    iterations=6000,
    task_type="GPU",
    devices='0',
    cat_features=cat_cols
)
```

```python
start = time.time()
model_catboost_clf.fit(X_train, y_train)
end = time.time()
print("Время обучения:", end - start)
```

```python
catboost_pred_train = model_catboost_clf.predict(X_train.copy())
catboost_pred_f1_score_train = f1_score(catboost_pred_train, y_train.copy(), average='macro')
print(f'F1-score train: {catboost_pred_f1_score_train }')

catboost_pred_test = model_catboost_clf.predict(X_valid.copy())
catboost_pred_f1_score_test = f1_score(catboost_pred_test, y_valid.copy(), average='macro')
print(f'F1-score test: {catboost_pred_f1_score_test}')
```
