```python
def mserror(X, w1, w0, y):
    y_pred = w1 * X[:, 0] + w0 # w1 - наклон, w0 - сдвиг
    return np.sum((y - y_pred) ** 2) / len(y_pred)

# Функция градиента
def gr_mserror(X, w1, w0, y):
    y_pred = w1 * X[:, 0] + w0
    return np.array([2/len(X)*np.sum((y - y_pred)) * (-1),       # Сдвиг
                     2/len(X)*np.sum((y - y_pred) * (-X[:, 0]))]) # Наклон
```

```python
X = np.array(X, type(float))
y = np.array(y, type(float))
```

```python
def learn():
    eps = 0.0001
    w1 = 0
    w0 = 0
    
    lr = 0.001
    
    next_w1 = w1
    next_w0 = w0
    
    n = 100000
    
    for i in range(n):
        cur_w1 = next_w1
        cur_w0 = next_w0
    
        next_w0 = cur_w0 - lr * gr_mserror(X, cur_w1, cur_w0, y)[0]
        next_w1 = cur_w1 - lr * gr_mserror(X, cur_w1, cur_w0, y)[1]
    
        print(f"Итерация: {i}")
        print(f"Текущая точка {cur_w1, cur_w0} | Следующая точка {next_w1, next_w0}")
        print(f"MSE {mserror(X, cur_w1, cur_w0, y)}")
        print("-"*50)
    
        if (abs(cur_w1 - next_w1) <= eps) and (abs(cur_w0 - next_w0) <= eps):
            return [i, cur_w0, cur_w1, next_w0, next_w1, mserror(X, cur_w1, cur_w0, y)]
```

```python
learning_res = learn()
```

```python
groups = df.groupby(by='region').groups

# Добавляем конкретные группы
bmi = df['bmi']
southwest = bmi[groups['southwest']]
southeast = bmi[groups['southeast']]
northwest = bmi[groups['northwest']]
northeast = bmi[groups['northeast']]
```

```python
import scipy.stats as sts
```

```python
sts.f_oneway(southwest, southeast, northwest, northeast)
```

p-значение меньше 0.05, следовательно группы значительно отличаются друг от друга


```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
model = ols('bmi ~ region', data=df[['region', 'bmi']]).fit()
sm.stats.anova_lm(model, typ=2)
```

Вычисленное p-значение совпало с вычисленным в предыдущем пункте.
Следовательно, мы можем сделать тот же вывод о различии групп.

```python
# С помощью t критерия Стьюдента перебрать все пары.
# Определить поправку Бонферрони. Сделать выводы.
region_pairs = []
regions = ['southwest', 'southeast', 'northwest', 'northeast']

for i, region_i in enumerate(regions):
    for region_j in regions[i + 1:]:
        region_pairs.append((region_i, region_j))

region_pairs
```

```python
m = len(region_pairs)
critical_value = 0.05 / m

result_cols = ['region_1', 'region_2', 'p_value', 'critical_p_value', 'result']
result = []
for region_1, region_2 in region_pairs:
    row = []
    row.extend([region_1, region_2])
    result_t_student = sts.ttest_ind(bmi[groups[region_1]],  bmi[groups[region_2]])
    pvalue = result_t_student.pvalue
    row.append(pvalue)
    row.append(critical_value)
    row.append('Группы значимо не отличаются' if pvalue > critical_value else 'Группы значимо отличаются')
    result.append(row)

pd.DataFrame(result, columns=result_cols)
```

Группы `northwest` и `northeast` не отличаются значимо.
Остальные группы между собой значимо отличаются.

```python
# Выполнить пост-хок тесты Тьюки и построить график.
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukeys = pairwise_tukeyhsd(endog=df['bmi'], groups=df['region'], alpha = 0.05)
tukeys.plot_simultaneous()
plt.vlines(x=np.mean(df['bmi']), ymin=-0.5, ymax=3.5)
tukeys.summary()
```

Пост-хок тест Тьюки дал аналогичный результат, что и расчёты в предыдущем пункте.


```python
# Выполнить двухфакторный ANOVA тест, 
# чтобы проверить влияние региона и пола на индекс массы тела (BMI),
# используя функцию anova_lm() из библиотеки statsmodels.

model = ols('bmi ~ C(region) + C(sex) + C(region):C(sex)', data = df[['region', 'bmi', 'sex']]).fit()
anova_result = sm.stats.anova_lm(model, type = 2)
print(anova_result)
```

* Для региона p-значение меньше 0.05, значит он влияет на ИМТ
* Для пола p-значение больше 0.05, значит он не влияет на ИМТ
* Для взаимодействия региона и пола p-значение больше 0.05, значит оно не влияет на ИМТ

```python
# Выполнить пост-хок тесты Тьюки и построить график.

df['combination'] = df['region'] + " / " + df['sex']

tukey = pairwise_tukeyhsd(endog = df['bmi'], groups = df['combination'], alpha = 0.05)
tukey.plot_simultaneous()
tukey.summary()
```

* Внутри регионов нет особых различий ИМТ по полу.
* Общая картина распределений осталась такой же, как и для однофакторного анализа регионов
